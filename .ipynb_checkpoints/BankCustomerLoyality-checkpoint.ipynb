### Artificial Neural Network

Part 1 - Data Preprocessing

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cd Dataset
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
X
```

```python
y
```

```python
#Create dummy variables
geography = pd.get_dummies(X["Geography"],drop_first=True)
gender = pd.get_dummies(X["Gender"],drop_first=True)
```

```python
geography
```

```python
gender
```

```python
# Concatenate the Data Frames
X = pd.concat([X,geography,gender],axis=1)
```

```python
X
```

```python
# Drop Unnecessary columns
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

Part 2 - Now let's make the ANN!

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,PReLU,LeakyReLU,ELU
```

```python
# Initialising the ANN

classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(6, init = 'he_uniform',activation='relu',input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(6, init = 'he_uniform', activation='relu'))

# Adding the output layer
classifier.add(1, init="glorot_uniform", activation='sigmoid')

# Compiling the ANN
classifier.compile(optimizer='Adamax', loss='binary_crossentropy', metrics=['accuracy'])

# Fitting the ANN to the Training set
model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs = 100)
```
